```python
from pydub import AudioSegment

audio_file = "audio.wav"
start_times = [1.469384, 8.625907, 20.272930]  # Specify the starting times of each segment in seconds
end_times = [6.796706, 17.095578, 26.204054]  # Specify the ending times of each segment in seconds

sound_file = AudioSegment.from_wav("audio.wav")

for i, (start_time, end_time) in enumerate(zip(start_times, end_times)):
    start_time_ms = int(start_time * 1000)  # Convert start time to milliseconds
    end_time_ms = int(end_time * 1000)  # Convert end time to milliseconds

    segment = sound_file[start_time_ms:end_time_ms]
    out_file = f"segment{i+1}.wav"
    print("Exporting", out_file)
    segment.export(out_file, format="wav")

```
